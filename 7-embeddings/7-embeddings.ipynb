```python
from sklearn import manifold
```

```python
highlited_vectors = [
    wv_from_text['szkoda::noun'],
    wv_from_text['strata::noun'],
    wv_from_text['uszczerbek::noun'],
    wv_from_text['szkoda_majątkowa::noun'],
#     wv_from_text['uszczerbek::noun'] + wv_from_text['na::prep'] +wv_from_text['zdrowie::noun'],
    wv_from_text['Krzywda::noun'],
    wv_from_text['niesprawiedliwość::noun'],
    wv_from_text['nieszczęście::noun']
]
```

```python
keys = list(wv_from_text.vocab.keys())
```

```python
random_keys = np.random.choice(keys[:800000],  1000)
```

```python
random_vectors = list(map(lambda key:  wv_from_text[key], random_keys))
```

```python
all_vectors = highlited_vectors + random_vectors
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(all_vectors)
```

```python
import matplotlib.pyplot as plt
```

```python
h = len(highlited_vectors)
```

```python
plt.figure(figsize=(18,9))
plt.scatter(X_embedded[h+1:, 0], X_embedded[h+1:, 1])
plt.scatter(X_embedded[:h, 0], X_embedded[:h, 1], c='r')
plt.show()
```
